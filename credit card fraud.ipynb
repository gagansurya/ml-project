```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

distribution of normal and fraud transactions

```python
df['Class'].value_counts()
```

```python
#sepeate data 
normal = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
normal.shape
```

```python
fraud.shape
```

building new normal dataset containing similar distributions of normal and fraud transactions

```python
normal_sample = normal.sample(n=492)
```

```python
#concatinating the two dataset
new_dataset = pd.concat([normal_sample,fraud],axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.shape
```

spliting the new_dataset into x and y

```python
x = new_dataset.drop(columns = 'Class',axis = 1)
y = new_dataset['Class']
```

```python
x
```

```python
y
```

splitting dataset into training and testing

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2, stratify=y,random_state=42)
```

```python
print(x_train,x_test,y_train,y_test)
```

model training and testing

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

accuracy score

```python
y_predict = model.predict(x_test)
```

```python
acc = accuracy_score(y_test,y_predict)
```

```python
acc
```

so we got accuracy of 92% by training a model using logistic regression
